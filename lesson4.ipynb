```python
import pandas as pd
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('house_pricing_dataset.csv')

print(df)
```

```python
plt.hist(df['price'], bins=400)
plt.xscale("log")
plt.xlabel('Prices')
plt.ylabel('Houses')
plt.title('Distribution of House Prices')
plt.show()
```

```python
print(df.describe())
```

```python
plt.hist(df.groupby('city')['price'].mean().sort_values(ascending=False).head(10));
```

```python
plt.figure(figsize=(16,10))
plt.scatter(df['date'],df['price'])
plt.xticks(df['date'], rotation='vertical')
plt.show()
```

```python
plt.hist(df.groupby('city')['bedrooms'].max().sort_values(ascending=False).head(10));
```

```python
oldhouse = df['yr_built'].head(10)
old = oldhouse.min()
cities = df['city'].head(10)
# print(cities)
plt.figure(figsize=(16,10))
plt.plot(cities, [old]*10, color= 'pink')
plt.bar(cities , oldhouse)
plt.show()
```

```python
# plt.figure(figsize=(51,10))
plt.bar(df['yr_built'],df['yr_renovated'])
plt.show()
```

```python
cities = df['city'].head(10)
basements = df['sqft_basement'].mean()
plt.figure(figsize=(16,10))
plt.bar(cities , basements)
plt.show()
```
