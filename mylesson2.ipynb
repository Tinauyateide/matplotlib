```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('anscombe_quartet_data.csv')
print(df)
```

```python
print(df.describe())
```

```python
x=np.array([4,15])
y=np.array([4,11])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('X1 againt Y1')
plt.scatter(df['x1'],df['y1'], color='green')
plt.xlabel('x1')
plt.ylabel('y1')
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title('X2 against Y2')
plt.scatter(df['x2'], df['y2'], color='red')
plt.xlabel('x2')
plt.ylabel('y2')
plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title('X3 against Y3')
plt.scatter(df['x3'], df['y3'], color='yellow')
plt.xlabel('x3')
plt.ylabel('y3')
plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title('X4 against Y4')
plt.scatter(df['x4'],df['y4'], color='black')
plt.xlabel('x4')
plt.ylabel('y4')
plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show() 
```

```python
x=np.array([4,15])
y=np.array([4,11])
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

ax1.plot(x,y)
ax1.scatter(df['x1'], df['y1'], color='grey')
ax1.set_title('X1 against Y1')
plt.xlabel('x1')
plt.ylabel('y1')
ax2.plot(x,y)
ax2.scatter(df['x2'],df['y2'], color='pink')
ax2.set_title('X2 against Y2')
plt.xlabel('x2')
plt.ylabel('y2')
ax3.plot(x,y)
ax3.scatter(df['x3'], df['y3'], color='yellow')
ax3.set_title('X3 against Y3')
plt.xlabel('x3')
plt.ylabel('y3')
ax4.plot(x,y)
ax4.scatter(df['x4'], df['y4'], color='violet')
ax4.set_title('X4 against Y4')
plt.xlabel('x4')
plt.ylabel('y4')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```

```python
x=np.array([4,15])
y=np.array([4,11])
fig = plt.figure(figsize=(10, 8))

ax5 = fig.add_subplot(221, title='X1 against Y1', xlabel='x1', ylabel='y1')
ax5.plot(x,y)
ax5.scatter(df['x1'], df['y1'], color='pink')
ax6 = fig.add_subplot(222, title='X2 against Y2', xlabel='x2', ylabel='y2' )
ax6.plot(x,y)
ax6.scatter(df['x2'], df['y2'], color='blue')
ax7 = fig.add_subplot(223, title='X3 against Y3', xlabel='x3', ylabel='y3')
ax7.plot(x,y)
ax7.scatter(df['x3'], df['y3'], color='gold')
ax8 = fig.add_subplot(224, title='X4 against Y4', xlabel='x4', ylabel='y4')
ax8.plot(x,y)
ax8.scatter(df['x4'], df['y4'], color='purple')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```
