```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
print(df)
```

```python
plt.bar(df['month_number'] - 0.2, df['moisturizer'], 0.4, color='orange', label='Moisturizer')
plt.bar(df['month_number'] + 0.2, df['facecream'], 0.4, color='blue', label='Facecream')
plt.title('My Sales Data on Facial Products')
plt.xlabel('Month')
plt.ylabel('Sales_Data')
plt.legend()
plt.savefig('mysalesdata.png')
plt.show()
```

```python
monthList  = df ['month_number']
bathingsoapSalesData   = df ['bathingsoap']
yerr = df['bathingsoap']/10000
# plt.bar(monthList, bathingsoapSalesData)
plt.bar(range(len(monthList)), bathingsoapSalesData, xerr=yerr, capsize=10, color='orange')
plt.xlabel('Month')
plt.ylabel('Sales_data')
plt.title('bathingsoap sales data')
plt.show()
```

```python
monthList  = df ['month_number'].tolist()

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%d%%')
plt.legend(loc='center left')
plt.title('Sales data')
plt.show()
```
